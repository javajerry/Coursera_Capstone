### import libraries

```python
import PyPDF2 as pyPdf
import os
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image #converting images into arrays

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python

# open the file and read it into a variable           
content=open('/Users/sumit/Downloads/JUSTICE_POEM.txt','r').read()
```

```python
stopwords=set(STOPWORDS)
stopwords.add('Till')
stopwords.add('Don')
stopwords.add('ill')
stopwords.add('will')
stopwords.add('longs')
stopwords.add('len')
stopwords.add('Whether')
stopwords.add('round')
#stopwords.add('ll')


#  instantiate a word cloud object
dream_wc = WordCloud(background_color='white',
                     max_words=2000,
                     stopwords=stopwords
                    )

# generate  the word cloud
dream_wc.generate(content)
```

```python
# display the word cloud

fig=plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18)# set height

plt.imshow(dream_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# save mask to mlk_mask
mlk_mask=np.array(Image.open('/Users/sumit/Downloads/chel-neha2.PNG'))

fig=plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18)# set height

plt.imshow(mlk_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#  instantiate a word cloud object
dream_wc = WordCloud(background_color='white',
                     max_words=2000,
                     mask=mlk_mask,
                     contour_width=1,
                     contour_color='steelblue',
                     margin=5,
                     stopwords=stopwords
                     
                    )

# generate  the word cloud
dream_wc.generate(content)

# display the word cloud

fig=plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18)# set height

plt.imshow(dream_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
fig.savefig('/Users/sumitchhabra/Downloads/chel-neha-wordcloud.PNG')
```
