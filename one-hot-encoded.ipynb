```python
import pandas as pd
df = pd.DataFrame({'color' : ['red','red','blue','red','green']})
df
```

```python
one_hot_df = pd.get_dummies(df[['color']])
one_hot_df
```

```python
one_hot_df['color']=df['color']
one_hot_df
```

```python
# move color column to the first column
fixed_columns = [one_hot_df.columns[-1]] + list(one_hot_df.columns[:-1])
#print("Fixed Columns:", fixed_columns)
one_hot_df = one_hot_df[fixed_columns]
one_hot_df
```

```python
one_hot_df.groupby('color').count()
```

```python
grouped = one_hot_df.groupby('color').mean().reset_index()
grouped
```
